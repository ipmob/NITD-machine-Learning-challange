```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('data/train_file.csv')
dataset
```

```python
data=dataset.drop(['ID','UsageClass','CheckoutYear','CheckoutMonth','Creator','CheckoutType','Checkouts','PublicationYear'],axis=1)
data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data.info()
type(data['MaterialType'])
```

```python
data['Title']=data['Title'].astype('str',inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 0.1)
```

```python
Xtest=[]
Xtrain=[]
for i in range(len(xTest)):
    Xtest.append(str(xTest.iloc[i]['Title']))
    Xtest[i]+=(' '+str(xTest.iloc[i]['Subjects']))
    Xtest[i]+=(' '+str(xTest.iloc[i]['Publisher']))
for i in range(len(xTrain)):
    Xtrain.append(str(xTrain.iloc[i]['Title']))
    Xtrain[i]+=(' '+str(xTrain.iloc[i]['Subjects']))
    Xtrain[i]+=(' '+str(xTrain.iloc[i]['Publisher']))
```

```python

vectorizer=TfidfVectorizer()
Xtrain = vectorizer.fit_transform(Xtrain)
Xtest = vectorizer.transform(Xtest)
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression(random_state=0,solver='lbfgs').fit(Xtrain,yTrain)
```

```python
pred=clf.predict(Xtest)
clf.score(Xtest,yTest)
```

```python
testset=pd.read_csv('data/test_file.csv')
testset
test=testset.drop(['ID','UsageClass','CheckoutYear','CheckoutMonth','Creator','CheckoutType','Checkouts','PublicationYear'],axis=1)
test
```

```python
testing=[]
for i in range(len(test)):
    testing.append(str(test.iloc[i]['Title']))
    testing[i]+=(' '+str(test.iloc[i]['Subjects']))
    testing[i]+=(' '+str(test.iloc[i]['Publisher']))
testing
```
